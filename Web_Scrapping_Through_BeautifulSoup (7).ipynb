```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Function to fetch header tags from a URL
def get_header_tags(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    header_tags = [header.text.strip() 
                   
for header in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])]
    return header_tags
```

```python
# Wikipedia homepage URL
wiki_url = 'https://en.wikipedia.org/wiki/Main_Page'

# Get header tags
header_tags = get_header_tags(wiki_url)
```

```python
# Create a DataFrame
df = pd.DataFrame({'Header Tags': header_tags})
print(df)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to fetch data from the given URL
def get_presidents_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extracting data using BeautifulSoup
    presidents_data = []
    for row in soup.find_all('tr')[1:]:  # Skip the header row
        columns = row.find_all('td')
        name = columns[0].text.strip()
        term_of_office = columns[1].text.strip()
        presidents_data.append({'Name': name, 'Term of Office': term_of_office})
    
    return presidents_data

# URL of the webpage containing information about former presidents of India
presidents_url = 'https://presidentofindia.nic.in/former-presidents.htm'

# Get data about former presidents
presidents_data = get_presidents_data(presidents_url)

# Create a DataFrame
df = pd.DataFrame(presidents_data)

# Display the DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_odi_teams():
    url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    teams_data = []

    teams_table = soup.find('table', {'class': 'table'})
    rows = teams_table.find('tbody').find_all('tr')[:10]

    for row in rows:
        columns = row.find_all('td')
        team = columns[1].text.strip()
        matches = columns[2].text
        points = columns[3].text
        rating = columns[4].text.strip()

        teams_data.append({'Team': team, 'Matches': matches, 'Points': points, 'Rating': rating})

    return pd.DataFrame(teams_data)

```

```python

def scrape_odi_batsmen():
    url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    batsmen_data = []

    batsmen_table = soup.find('table', {'class': 'table'})
    rows = batsmen_table.find('tbody').find_all('tr')[:10]

    for row in rows:
        columns = row.find_all('td')
        player = columns[1].text.strip()
        team = columns[2].text.strip()
        rating = columns[3].text.strip()

        batsmen_data.append({'Player': player, 'Team': team, 'Rating': rating})

    return pd.DataFrame(batsmen_data)

```

```python
def scrape_odi_bowlers():
    url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    bowlers_data = []

    bowlers_table = soup.find('table', {'class': 'table'})
    rows = bowlers_table.find('tbody').find_all('tr')[:10]

    for row in rows:
        columns = row.find_all('td')
        player = columns[1].text.strip()
        team = columns[2].text.strip()
        rating = columns[3].text.strip()

        bowlers_data.append({'Player': player, 'Team': team, 'Rating': rating})

    return pd.DataFrame(bowlers_data)

```

```python
if __name__ == "__main__":
    
    # a) Top 10 ODI teams
    odi_teams_df = scrape_odi_teams()
    print("Top 10 ODI Teams:")
    print(odi_teams_df)
    print()

    # b) Top 10 ODI Batsmen
    odi_batsmen_df = scrape_odi_batsmen()
    print("Top 10 ODI Batsmen:")
    print(odi_batsmen_df)
    print()

    # c) Top 10 ODI Bowlers
    odi_bowlers_df = scrape_odi_bowlers()
    print("Top 10 ODI Bowlers:")
    print(odi_bowlers_df)
 
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
```

```python
    paper_titles = []
    authors_list = []
    published_dates = []
    paper_urls = []

    
    articles = soup.find_all("div", class_="pod-listing-header")
    for article in articles:
        
        # Extracting paper title
        title = article.find("a", class_="pod-listing-header__link").text.strip()
        paper_titles.append(title)

        # Extracting authors
        authors = article.find("div", class_="pod-listing-header__authors").text.strip()
        authors_list.append(authors)

        # Extracting published date
        date = article.find("div", class_="pod-listing-header__date").text.strip()
        published_dates.append(date)

        # Extracting paper URL
        paper_url = article.find("a", class_="pod-listing-header__link")["href"]
        paper_urls.append(paper_url)

    # Creating a DataFrame
    data = {
        'Paper Title': paper_titles,
        'Authors': authors_list,
        'Published Date': published_dates,
        'Paper URL': paper_urls
    }

    df = pd.DataFrame(data)
    print(df)

```
